## Grouping the variable :

### If we add all these dummy variable in our dataframe , we would end up the dataset containing nearly 40 columns. This become a huge dataset with 40 columns and 700 observations in this situation , when we have so many variable of the same type , we should always consider the posibility of grouping these variables.

### Note :-

### GROUP = CLASS

### A group can also be called class.

### Group these variables(dummy variable) :-

### Re-oragnizing a certain type of variables into group in a regression analysis , This process is also called Classification.

### Now,  There are various technique that decide on how to group these variable in our dataset.

### For our example , we focus on qualitative analysis because it seems most appropriate.

### We have description table that show all possible reason for absence which indicate numbers , and then , we observe that :-

### 2. The second class could inculde Reason from 15 to 17 because they all related to pregency and giving birth.

## Analyzing Absenteeism Time in Hours column

### By looking at the data of this column we found that this column represents the number of hours an employee away from work , as we know it is the regression analysis because our task is to predict the number of hours which employee is away from work.

### But by predicting only the number of hours doesn't seem so good , as we don't what is limit we decide that someone taking the excess of absenteeism who's not. 

### So my making this problem simple to the point of view of classification purpose like if we predict which employee excess the absenteeism and which employee not that will make easy for our analysis and interpretation.

### So we have to convert our tagert or dependent variable into somehow categorical format so we apply any classification algorithm and get easily interpretation result who will excess the absenteeism and who will not

### There were two classes
### 1. Employee who Excessively absent
### 2. Employee who Moderately  absent

### How is this possible to do so ?

### We will use a methodolgy which is tiny bit naive but numerically stable

### We will take the median value of the 'Absenteeism Time in Hours' and use it as a cut-off line.

### Everything below the median considered normal and Everything above the median would be excessive

### Now we have to find that the median of 'Absenteeism Time in Hours' is 3.0 and it's datatype is float

### Now we will apply the following reasoning

### There were two classes
### 1. Everyone that has been absent for more than 3 Hours will be considered as excessively absent
### 2. Everyone that has been absent for less than 3 Hours will be considered as moderately absent

## Feature Scaling 

### The most common problem in working with numerical data is difference in magnitudes. To fix this issue is Standardization also known as Feature Scaling.

### Feature Scaling(Standardization) is the process of transforming the data we are working with into standard scale.

### Mathematically ,

### Standarad value = original variable - mean of original variable / standard deviation of original variable

### This translate to substracting the mean and dividing by the standard deviation.

### In this way regardless of the dataset we will always obtains a distribution with a mean of zero and a standard deviation of one.

### In our dataset we cannot standarized the whole data because in our dataset we have dummy variable and if we standardize the dummy variable we loose the whole interpretability.

### We create a new variable from which we include only our input variable including dummy variable.

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs
```

### This is the Custom scaler based on the standard scaler from sklearn , 

### However when we declare the scaler object there's an extra agrument columns to scale

### So , our custom scale will not standardized all inputs but only the ones we choose , In this way we will be able to preserve the dummy variable untouched

### We have CustomScaler Class there is no different  than the standard scalar in the way it works 

```python
from sklearn.base import BaseEstimator , TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator,TransformerMixin):
    
    def __init__(self,columns,copy=True,with_mean=True,with_std=True):
        self.scaler = StandardScaler(copy,with_mean,with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    def fit(self,X,y=None):
        self.scaler.fit(X[self.columns],y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self,X,y=None,copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]),columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled,X_scaled],axis=1)[init_col_order]
```
